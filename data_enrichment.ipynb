# Сбор и обогащение датасета

```python
import pandas as pd
```

## Сбор данных по заказам Olist (датасет Kaggle)

За основу мы взяли датасет с Kaggle: [Brazilian E-Commerce Public Dataset by Olist](https://www.kaggle.com/datasets/olistbr/brazilian-ecommerce)

```python
orders = pd.read_csv(f'data/olist/olist_orders_dataset.csv')
orders
```

```python
orders = orders[orders['order_status'] == 'delivered'].copy()
orders
```

```python
date_cols = ['order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col])
```

```python
order_items = pd.read_csv(f'data/olist/olist_order_items_dataset.csv')
order_items
```

```python
items_agg = order_items.groupby('order_id').agg(
    seller_id = ('seller_id', 'first'),
    product_id = ('product_id', 'first'),
    n_items = ('order_item_id', 'count'),
    total_price = ('price', 'sum'),
    total_freight = ('freight_value', 'sum'),
    avg_price = ('price', 'mean')
).reset_index()

items_agg
```

```python
payments = pd.read_csv(f'data/olist/olist_order_payments_dataset.csv')
payments
```

```python
pay_agg = payments.groupby('order_id').agg(
    payment_type = ('payment_type', 'first'),
    payment_installments = ('payment_installments', 'max'),
    payment_value = ('payment_value', 'sum')
).reset_index()
pay_agg
```

```python
reviews = pd.read_csv(f'data/olist/olist_order_reviews_dataset.csv')
reviews
```

```python
reviews_agg = reviews.groupby('order_id')['review_score'].mean().reset_index()
reviews_agg
```

```python
customers = pd.read_csv(f'data/olist/olist_customers_dataset.csv')
customers
```

```python
sellers = pd.read_csv(f'data/olist/olist_sellers_dataset.csv')
sellers
```

```python
products = pd.read_csv(f'data/olist/olist_products_dataset.csv')
products
```

```python
geolocation = pd.read_csv(f'data/olist/olist_geolocation_dataset.csv')
geolocation
```

```python
geo_agg = geolocation.groupby('geolocation_zip_code_prefix').agg(
    lat = ('geolocation_lat', 'mean'),
    lng = ('geolocation_lng', 'mean')
).reset_index()
```

```python
geo_agg_customers = geo_agg.rename(columns={'geolocation_zip_code_prefix': 'customer_zip_code_prefix', 'lat': 'customer_lat', 'lng': 'customer_lng'})
customers_geo = pd.merge(customers, geo_agg_customers, on='customer_zip_code_prefix', how='left')
customers_geo
```

```python
geo_agg_sellers = geo_agg.rename(columns={'geolocation_zip_code_prefix': 'seller_zip_code_prefix', 'lat': 'seller_lat', 'lng': 'seller_lng'})
sellers_geo = pd.merge(sellers, geo_agg_sellers, on='seller_zip_code_prefix', how='left')
sellers_geo
```

```python
category_map = pd.read_csv(f'data/olist/product_category_name_translation.csv')
category_map
```

```python
products = pd.merge(products, category_map, on='product_category_name', how='left')
products
```

```python
df = orders.copy()

df = pd.merge(df, items_agg, on='order_id', how='left')
df = pd.merge(df, pay_agg, on='order_id', how='left')
df = pd.merge(df, reviews_agg, on='order_id', how='left')
df = pd.merge(df, customers_geo[['customer_id', 'customer_unique_id', 'customer_state', 'customer_city', 'customer_lat', 'customer_lng']], on='customer_id', how='left')
df = pd.merge(df, sellers_geo[['seller_id', 'seller_state', 'seller_city', 'seller_lat', 'seller_lng']], on='seller_id', how='left')
df = pd.merge(df, products[['product_id', 'product_category_name_english', 'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm', 'product_photos_qty']], on='product_id', how='left')

df
```

```python
df['delay_days'] = (df['order_delivered_customer_date'] - df['order_estimated_delivery_date']).dt.total_seconds() / 86400
df['is_delayed'] = (df['delay_days'] > 0).astype(int)

df['seller_processing_days'] = (df['order_delivered_carrier_date'] - df['order_approved_at']).dt.total_seconds() / 86400
df['transit_days'] = (df['order_delivered_customer_date'] - df['order_delivered_carrier_date']).dt.total_seconds() / 86400

df['order_weekday'] = df['order_purchase_timestamp'].dt.dayofweek
df['order_month'] = df['order_purchase_timestamp'].dt.month
df['order_year'] = df['order_purchase_timestamp'].dt.year
df['order_hour'] = df['order_purchase_timestamp'].dt.hour

df['freight_ratio'] = df['total_freight'] / (df['total_price'] + 1e-9)

df['product_volume_cm3'] = (df['product_length_cm'] * df['product_height_cm'] * df['product_width_cm'])

df['low_review'] = (df['review_score'] <= 3).astype(int)

df
```

```python
import numpy as np

def get_distance(lat1, lon1, lat2, lon2):
    R = 6371
    lat1 = np.radians(lat1)
    lon1 = np.radians(lon1)
    lat2 = np.radians(lat2)
    lon2 = np.radians(lon2)

    a = np.sin((lat2 - lat1) / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin((lon2 - lon1) / 2)**2
    return 2 * R * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

df['distance_km'] = get_distance(df['seller_lat'].values, df['seller_lng'].values, df['customer_lat'].values, df['customer_lng'].values)
df
```

## Обогащение данными по штатам Бразилии (IBGE)

```python
import requests

r = requests.get('https://servicodados.ibge.gov.br/api/v1/localidades/estados', timeout=15)
raw = r.json()

states_df = pd.DataFrame([{'state_uf': s['sigla'], 'state_name': s['nome'], 'region': s['regiao']['nome'], 'region_sigla': s['regiao']['sigla'], 'ibge_id': str(s['id'])} for s in raw])

r = requests.get('https://servicodados.ibge.gov.br/api/v3/agregados/202/periodos/2010/variaveis/93?localidades=N3[all]', timeout=15)
pop_series = r.json()[0]['resultados'][0]['series']
pop_dict = {s['localidade']['id']: int(list(s['serie'].values())[0]) for s in pop_series}

states_df['population'] = states_df['ibge_id'].map(pop_dict)
states_df = states_df.drop(columns='ibge_id')
states_df
```

```python
state_extra = pd.read_csv('data/state_extra.csv')
state_extra
```

```python
states_full = pd.merge(states_df, state_extra, on='state_uf', how='left')
states_full['gdp_per_capita'] = (states_full['gdp_brl_billion'] * 1e9 / states_full['population']).round(0)
states_full['population_density'] = (states_full['population'] / (states_full['area_km2_thousands'] * 1000)).round(2)

states_full
```

```python
states_customer = states_full[['state_uf', 'region', 'region_sigla', 'population', 'gdp_per_capita', 'hdi', 'urban_pct', 'population_density']].rename(columns=lambda c: f'customer_{c}' if c != 'state_uf' else 'customer_state')
states_seller = states_full[['state_uf', 'region', 'gdp_per_capita', 'hdi', 'urban_pct']].rename(columns=lambda c: f'seller_{c}' if c != 'state_uf' else 'seller_state')

df = pd.merge(df, states_customer, on='customer_state', how='left')
df = pd.merge(df, states_seller, on='seller_state', how='left')

df['hdi_diff'] = df['customer_hdi'] - df['seller_hdi']
df
```

## Обогащение данными погоды (Open-Meteo)

```python
import openmeteo_requests
import requests_cache
from retry_requests import retry

cache_session = requests_cache.CachedSession('.openmeteo_cache', expire_after=-1)
openmeteo = openmeteo_requests.Client(session=retry(cache_session, retries=3, backoff_factor=0.5))

state_coords = pd.read_csv('data/state_coords.csv')
state_coords = dict(zip(state_coords['state_uf'], zip(state_coords['lat'], state_coords['lon'])))
```

```python
weather_frames = []

for state_uf, (lat, lon) in state_coords.items():
    resp = openmeteo.weather_api('https://archive-api.open-meteo.com/v1/archive',
        params={'latitude': lat, 'longitude': lon, 'start_date': '2016-09-01', 'end_date': '2018-09-30', 'daily': ['precipitation_sum', 'temperature_2m_max', 'temperature_2m_min'], 'timezone': 'America/Sao_Paulo'})[0].Daily()

    dates = pd.date_range(start=pd.Timestamp(resp.Time(), unit='s', tz='America/Sao_Paulo').date(), periods=resp.Variables(0).ValuesAsNumpy().shape[0], freq='D').date

    wdf = pd.DataFrame({
        'date': dates,
        'precipitation_mm': resp.Variables(0).ValuesAsNumpy().round(2),
        'temp_avg_c': ((resp.Variables(1).ValuesAsNumpy() + resp.Variables(2).ValuesAsNumpy()) / 2).round(1),
        'heavy_rain': (resp.Variables(0).ValuesAsNumpy() > 20).astype(int),
    })

    wdf['state_uf'] = state_uf
    weather_frames.append(wdf)

weather_all = pd.concat(weather_frames, ignore_index=True)
weather_all
```

```python
df['purchase_date'] = df['order_purchase_timestamp'].dt.date
df['delivery_date'] = df['order_delivered_customer_date'].dt.date

weather_purchase = weather_all.rename(columns={'state_uf': 'customer_state', 'date': 'purchase_date', 'precipitation_mm': 'purchase_precipitation_mm', 'temp_avg_c': 'purchase_temp_avg_c', 'heavy_rain': 'purchase_heavy_rain'})
df = pd.merge(df, weather_purchase, on=['customer_state', 'purchase_date'], how='left')

weather_delivery = weather_all.rename(columns={'state_uf': 'customer_state', 'date': 'delivery_date', 'precipitation_mm': 'delivery_precipitation_mm', 'temp_avg_c': 'delivery_temp_avg_c', 'heavy_rain': 'delivery_heavy_rain'})
df = pd.merge(df, weather_delivery, on=['customer_state', 'delivery_date'], how='left')

df
```

## Обогащение праздниками Бразилии (python-holidays)

```python
import holidays

br_holidays = {}
for year in [2016, 2017, 2018]:
    br_holidays.update(holidays.Brazil(years=year))

holiday_dates = set(br_holidays.keys())
holiday_dates
```

```python
def days_to_next_holiday(date, holiday_dates):
    h = [h for h in holiday_dates if h > date]
    return (h[0] - date).days if h else None

purchase_dates = df['order_purchase_timestamp'].dt.date
delivery_dates = df['order_delivered_customer_date'].dt.date

df['purchase_is_holiday'] = purchase_dates.isin(holiday_dates).astype(int)
df['delivery_is_holiday'] = delivery_dates.isin(holiday_dates).astype(int)
df['days_to_next_holiday'] = purchase_dates.apply(lambda d: days_to_next_holiday(d, holiday_dates) if pd.notna(d) else None)

df
```
